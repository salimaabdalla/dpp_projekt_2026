```python
# Import moduls
import pandas as pd
import numpy as np

from math import nan

# Importing date transformation functions
import calendar
from datetime import datetime

pd.set_option('display.max_columns', None)
```

## Combining data from the different sources in one dataframe

```python
path_start = '../data/processed/'
files = ['1082ff_monatsbericht_suesswaren_schoko_kaffee_processed.csv', 
         '1082ff_erzeugerpreise_suesswaren_schoko_kaffee_processed.csv', 
         '1082_1083_einfuhrpreise_suesswarenundkaffee_processed.csv', 
         '01183ff_vpi_suesswaren_schoko_kaffee_processed.csv', 
         'A1014_17_18_monprod_suesswarenequi_kaffee_processed.csv', 
         'globalcacaoprice_fred_processed.csv', 
         'globalcoffeerobustaprice_fred_processed.csv']
```

Note that '../data/processed/1082_monatsbericht_suesswaren_timeseries_processed.csv' contains a subset of the information of '../data/processed/1082ff_monatsbericht_suesswaren_schoko_kaffee_processed.csv'

We load the dataframes in the different files. 

```python
list_of_df = []
for file in files:
    df = pd.read_csv(path_start + file, index_col = 'date', parse_dates = True)
    list_of_df.append(df)
df = pd.concat(list_of_df, axis=1)
```

```python
# transforming dates from first day of month to last day of month
# base for for loop:
date_list = list(df.index)
print(date_list[0])
date_list = [datetime.strptime(date, '%Y-%m-%d') for date in date_list]
input_dt = date_list[0]
print("The original date is:", input_dt.date())

# monthrange() to get the date range
res = calendar.monthrange(input_dt.year, input_dt.month)
day = res[1]
print(f"Last date of month is: {input_dt.year}-{input_dt.month}-{day}")
```

```python
# transforming all dates in the index to the last day of the month
new_index_list = []
for dates in date_list:
    res = calendar.monthrange(dates.year, dates.month)
    day = res[1]
    new_date = f"{dates.year}-{dates.month}-{day}"
    new_index_list.append(new_date)
new_index_list = [pd.to_datetime(date).strftime('%Y-%m-%d') for date in new_index_list]


# applying new index to dfs:
df.index = new_index_list
#print(dfs.index)
```

Load the share data and append them to the dataframe. 

```python
path = '../data/processed/shareprice_monthly.csv'
df_shares = pd.read_csv(path, index_col = 'date', skipfooter=1, parse_dates = True)

df = pd.concat([df, df_shares], axis=1)
df.index.name = 'date'
df
```

We reduce the spaces in the names of the columns.

```python
columns_without_space = []
for column in df.columns.values:
    columns_without_space.append(column.replace("  ", ""))
df.columns = columns_without_space
```

We save the dataframe. 

```python
df.to_csv('../data/processed/combined_data.csv')
```
