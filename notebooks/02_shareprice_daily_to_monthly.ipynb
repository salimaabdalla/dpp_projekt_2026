```python
from glob import glob

# alle Mondelez Dateien finden
file_list = glob('../data/raw/*_*_Mondelez.csv')
print(file_list)

df_Mondelez = pd.DataFrame()

for file in file_list:
        
    # CSV-Datei einlesen
    df = pd.read_csv(file, sep=';', decimal=',', skiprows=1)
    
    # Datum in datetime umwandeln und als Index setzen
    df['date'] = pd.to_datetime(df['Datum'], format='%d.%m.%y')
    df.set_index('date', inplace=True)

    # concating dataframes
    df_Mondelez = pd.concat([df_Mondelez, df], axis=0)

df_Mondelez.sort_index(inplace=True)
df_Mondelez
```

```python
# alle Ferrero Dateien finden
file_list = glob('../data/raw/*_*_Ferrero.csv')
print(file_list)

df_Ferrero = pd.DataFrame()

for file in file_list:
        
    # CSV-Datei einlesen
    df = pd.read_csv(file, sep=';', decimal=',', skiprows=1)
    
    # Datum in datetime umwandeln und als Index setzen
    df['date'] = pd.to_datetime(df['Datum'], format='%d.%m.%y')
    df.set_index('date', inplace=True)

    # concating dataframes
    df_Ferrero = pd.concat([df_Ferrero, df], axis=0)

df_Ferrero.sort_index(inplace=True)
df_Ferrero
```

```python
# alle LindtSpruengli Dateien finden
file_list = glob('../data/raw/*_*_LindtSpruengli.csv')
print(file_list)

df_LindtSpruengli = pd.DataFrame()
for file in file_list:
        
    # CSV-Datei einlesen
    df = pd.read_csv(file, sep=';', decimal=',', skiprows=1)
    
    # Datum in datetime umwandeln und als Index setzen
    df['date'] = pd.to_datetime(df['Datum'], format='%d.%m.%y')
    df.set_index('date', inplace=True)

    # concating dataframes
    df_LindtSpruengli = pd.concat([df_LindtSpruengli, df], axis=0)

df_LindtSpruengli.sort_index(inplace=True)
df_LindtSpruengli
```

```python
# monatliche Durchschnittswerte berechnen als letzter Tag im Monat
df_ferrero = df_Ferrero.loc[:, 'Schluss'].resample('ME').mean()
df_mondelez = df_Mondelez.loc[:, 'Schluss'].resample('ME').mean()
df_lindtspruengli = df_LindtSpruengli.loc[:, 'Schluss'].resample('ME').mean()

print(df_ferrero)
print(df_mondelez)  
print(df_lindtspruengli)
```

```python
# alle Produzenten nebeneinander einf√ºgen
df_shareprice_monthly = pd.concat([df_ferrero, df_mondelez, df_lindtspruengli], axis=1)
df_shareprice_monthly.columns = ['Ferrero', 'Mondelez', 'LindtSpruengli']
#df_shareprice_monthly.to_csv('C:\\Sali\\Organisation\\Bewerbung\\Arbeitsamt\\WeiterbildungDataScientist\\PortfolioProjekt\\modified\\shareprice_monthly.csv', sep=';', decimal=',')
df_shareprice_monthly.to_csv('../data/processed/shareprice_monthly.csv')
```
