```python
dataset_link = "declanmcalinden/time-series-uk-supermarket-data" # replace with your dataset link from Kaggle 
destination = "../data/raw"
dataset_name = dataset_link.split("/")[-1]

files = load_from_kaggle(
    dataset_link=dataset_link, 
    destination=destination,
    )
```

```python
df_aldi = pd.read_csv("/".join(["../data/raw/", dataset_name, files[0]]))
df_asda = pd.read_csv("/".join(["../data/raw/", dataset_name, files[1]]))
df_morrisons = pd.read_csv("/".join(["../data/raw/", dataset_name, files[2]]))
df_sains = pd.read_csv("/".join(["../data/raw/", dataset_name, files[3]]))
df_tesco = pd.read_csv("/".join(["../data/raw/", dataset_name, files[4]]))
df_aldi.head()
```

```python
df_list = []
for df in [df_aldi, df_asda, df_morrisons, df_sains, df_tesco]:
    df= df.dropna(subset = 'names')
    df_list.append(df.loc[df['names'].str.contains('choc', case=False), :])
df_choc = pd.concat([df_list[0], df_list[1], df_list[2], df_list[3], df_list[4]])
```

```python
df_1082 = pd.read_csv('../data/raw/1082_monatsbericht_suesswaren_timeseries.csv', skiprows =7,
            skipfooter = 4, sep = ';'
            )
df_1082
```

```python
df_1082.columns = ['Steller', 'Produkt', 'Betriebe', 'Betriebe_ep', 'Beschäftigte',
       'Beschäftigte_ep', 'Geleistete Arbeitsstunden', 'Geleistete Arbeitsstunden_ep',
       'Bruttolohn- und -gehaltssumme', 'Bruttolohn- und -gehaltssumme_ep', 'Umsatz', 'Umsatz_ep',
       'Inlandsumsatz', 'Inlandsumsatz_ep', 'Auslandsumsatz', 'Auslandsumsatz_ep',
       'Auslandsumsatz mit der Eurozone', 'Auslandsumsatz mit der Eurozone_ep',
       'Auslandsumsatz mit dem sonstigen Ausland', 'Auslandsumsatz mit dem sonstigen Ausland_ep']
```

```python
df_1082['month']=0
for k in [2, 198,  51, 149, 100, 296, 247]:
    for month in range(1,13): 
        for i in range(1,4):
            df_1082.iloc[k+i+(month-1)*4, -1] = month
for k in [2, 198,  51, 149, 100, 296, 247]:
    for month in range(0,12): 
        df_1082 =  df_1082.drop(k + month * 4,axis=0)
df_1082['year']=0
for k in range(7):
    for i in range(37):
        df_1082.iloc[1 + k* 37 + i, -1] = k + 2019
for k in range(7):
    df_1082 =df_1082.drop(1 + k*49,axis = 0)
```

```python
vielfachheiten = df_1082.iloc[0,:]
df_1082 =df_1082.drop(0,axis = 0)
```

```python
for col in ['Betriebe', 'Beschäftigte', 'Geleistete Arbeitsstunden', 'Bruttolohn- und -gehaltssumme', 
            'Inlandsumsatz', 'Auslandsumsatz', 'Auslandsumsatz mit der Eurozone', 
            'Auslandsumsatz mit dem sonstigen Ausland']:
    df_1082[col]=df_1082[col].replace('...', nan)
    df_1082[col]=df_1082[col].astype('float64')
```
