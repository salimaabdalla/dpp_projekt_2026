## 1082

```python
df= pd.read_csv('../data/raw/1082_1083_einfuhrpreise_suesswarenundkaffee.csv', 
                skiprows =6,
                skipfooter = 3, sep = ';'
                )

df
```

```python
for i in range(3, 171, 2):
    zahl = str(i)
    df = df.drop('Unnamed: '+zahl, axis=1)
df = df.transpose()
df.columns = ['Month', 'EP_Nahrungsmittel und Futtermittel', 'EP_Sonstige Nahrungsmittel (ohne Getränke)',
              'EP_Süßwaren (ohne Dauerbackwaren)', 'EP_Kaffee und Tee, Kaffee-Ersatz']
df = df.drop('Unnamed: 0', axis = 0)
df = df.drop('Unnamed: 1', axis = 0)
```

```python
df
```

```python
list_year = []
for i in range(7):
    list_year = list_year + np.full(12,i+2019).tolist()
df['Year'] = list_year
df['Day']=1
df['Month'] = df['Month'].replace({'Januar':1, 'Februar':2, 'März': 3, 'April':4, 'Mai':5, 
                                   'Juni': 6, 'Juli': 7, 'August': 8, 'September': 9, 'Oktober': 10, 'November': 11, 'Dezember': 12})

df['date']=pd.to_datetime(df[['Year', 'Month', 'Day']])
df = df.drop('Month', axis = 1)
df = df.drop('Year', axis = 1)
df = df.drop('Day', axis = 1)
df = df.set_index('date')
```

```python
df.to_csv('../data/processed/1082_1083_einfuhrpreise_suesswarenundkaffee_processed.csv')
```
